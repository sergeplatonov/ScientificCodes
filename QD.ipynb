```python
import scipy
import numpy
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
get_ipython().magic(u'matplotlib inline')
from matplotlib.colors import LogNorm
from __future__ import division
from scipy import ndimage
%config InlineBackend.figure_format = 'retina'
```

```python
exp1 = np.fromfile('D://stability.txt',dtype=float,count=-1, sep=" ")
exp1 = np.reshape(exp1, (85, -1))
exp2 = np.fromfile('D://stability2.txt',dtype=float,count=-1, sep=" ")
exp2 = np.reshape(exp2, (34, -1))
```

```python
#plt.imshow(exp1,cmap='Greys_r')
plt.plot(exp1[70]*-1)
plt.plot(exp1[:,40]*-1)
```

```python
from matplotlib import colors, cm
exp2=np.abs(exp2)
norm = colors.LogNorm(exp2.mean() + 0.5 * exp2.std(), exp2.max(), clip='True')
plt.imshow(exp2*1E9, cmap=cm.gray,vmin=1,vmax=20, norm=norm, origin="lower",aspect='auto')
#plt.imshow(exp2,cmap='Greys_r',aspect='auto')
```
